```python
medical_df = medical_df.sample(frac=1, random_state=42)
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(medical_df.drop("TYPE", axis=1), medical_df["TYPE"], test_size=0.2, random_state=42)
```

```python
def performance(trained_model):
    y_pred = trained_model.predict(x_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average="weighted")  # For multi-class, use 'weighted'
    recall = recall_score(y_test, y_pred, average="weighted")
    f1 = f1_score(y_test, y_pred, average="weighted")

    # Print the results
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
```

```python
# svc
svc_model = SVC(random_state=42)

svc_model.fit(x_train, y_train)
performance(svc_model)
```

```python
# linear svc
lsvc = LinearSVC(random_state=42)

lsvc.fit(x_train, y_train)
performance(lsvc)
```

```python
# RFC
rfc = RandomForestClassifier(random_state=42)

rfc.fit(x_train, y_train)
performance(rfc)
```

```python
# LogisticRegression
logiRe = LogisticRegression(random_state=42)

logiRe.fit(x_train, y_train)
performance(logiRe)
```

```python
# KNeighborsClassifier
from sklearn.metrics import rand_score


kNC = KNeighborsClassifier()

kNC.fit(x_train, y_train)
performance(kNC)
```

```python
# GaussianNB
gaus = GaussianNB()

gaus.fit(x_train, y_train)
performance(gaus)
```

```python
# DecisionTreeClassifier
dectree = DecisionTreeClassifier(random_state=42)

dectree.fit(x_train, y_train)
performance(dectree)
```
